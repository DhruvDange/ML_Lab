```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.cluster.hierarchy as shc
```

```python
data = pd.read_csv('datasets/advertising.csv')
data.head()
```

```python
data.drop(columns=['TV', 'radio', 'Unnamed: 0'], inplace=True)
```

```python
plt.figure(figsize=(10,7))
plt.title('AGNES')

dend = shc.dendrogram(shc.linkage(data, method='ward', metric='euclidean'))
```

```python
from sklearn.cluster import AgglomerativeClustering

clustering = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
clustering.fit(data)
clustering.labels_
```

```python
sns.scatterplot(x='newspaper', y='sales', data=data, hue=clustering.labels_)
```

```python
dend = shc.dendrogram(shc.linkage(data, method='ward',metric="euclidean"),orientation='bottom',distance_sort='ascending', show_leaf_counts=True)
```
